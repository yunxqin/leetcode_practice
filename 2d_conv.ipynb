```python
import numpy as np

# input_m=np.zeros((6,6))
input_m = np.random.randn(6, 6)

# kernel=np.zeros((3,3))
kernel = np.random.randint(0, 10, (3,3))
input_m
kernel
```

```python
print(input_m)
```

```python
import torch
import torch.nn.functional as F
```

```python
signal_t = torch.tensor(input_m, dtype=torch.float32)
kernel_t = torch.tensor(kernel, dtype=torch.float32)

signal_uq = signal_t.unsqueeze(0).unsqueeze(0)  # 变成 (1, 1, 3, 3)
kernel_uq = kernel_t.unsqueeze(0).unsqueeze(0)  # 变成 (1, 1, 2, 2)

# 使用 conv2d 进行二维卷积
result = F.conv2d(signal_uq, kernel_uq, padding=0)
print(result)
```

```python
from scipy.signal import convolve2d
input_m_2 = np.random.randint(0, 10, (6,6))
result = convolve2d(input_m_2, kernel, mode='valid')
print(result)
# convolution_result = np.convolve(input_m, kernel, mode='valid ')  # 'full' 返回完整的卷积结果
```

```python
size_k=kernel.shape[0]

output_m=[]

for row_ in range(0,len(input_m[0])-size_k+1,1):
    output_row=[]
    for line_ in range(0,len(input_m[:,:1])-size_k+1,1):
        #首先从输入中取和kernel-size一样的行数
        value_row=input_m[row_:row_+size_k]
        #从选出来的行数取和kernel size一样的列数
        value_line=value_row[:,line_:line_+size_k]
        sub_matrix=value_line
        print(sub_matrix)
        
        #计算每一个子矩阵和kernel的乘积,得到输出矩阵的第一行的单个值
        output_row_=0
        for i in range(0,size_k):
            for j in range(0,size_k):
                print(f"{sub_matrix[i][j]} * {kernel[i][j]}")
                output_row_+=sub_matrix[i][j]*kernel[i][j]

        print(output_row_)
        output_row.append(output_row_)
    print(output_row)
    output_m.append(output_row)
print(output_m)
```
